**SETS**

```python
#Set create
#Set is a collection which are unchanged, unordered,unindexed
S = {"hello","sves","cse"}
print(S)
print(S.add("add"))
```

```python
#no duplicates
S = {"Hello","Sves","CSE","Sves"}
print(S)
```

```python
#in sets 1 and True, 0 and False are considered as same
S = {"hello","Sves",True,1,0,False}
print (S)
```

```python
#length of the set
#len(setname)
S = {"Hello","slay",34,True,'64'}
print(len(S))
```

```python
#type
#type(setname)
S1={"hello",45,True,"67"}
print(type(S1))
S = {12,34,67}
print(type(S))
```

```python
#set constructor
#setname = set(("/"))
S = set(( "Hello","sves",65,8.0))
print(S)
```

```python
#for in sets
S = set(( "Hello","sves",65,8.0))
for i in S:
  print(i)
```

```python
#inserting elements in sets
S = set(( "Hello","sves",65,8.0))
S.add("New")
print(S)
```

```python
#updating elements in sets
S1={"Black","Red","Yellow"}
S = {"Green","Orange","Purple"}
S1.update(S)
print(S1)
```

```python
#Remove elements from set
S1={"Black","Red","Yellow"}
S1.remove("Yellow")
print(S1)
```

```python
#Discard #same as remove
S1={"Black","Red","Yellow"}
S1.discard("Yellow")
print(S1)
```

**TUPLES**

```python
#create tuple
#tuple_name = ("ele")
T = ("Yellow","Green","Orange")
print(T)
```

```python
#duplicates in tuples
T = ("Yellow","Green","Orange","Green")
print(T)
```

```python
#length in tuple
T = ("Yellow","Green","Orange")
print(len(T))
```

```python
#type in tuple
T = ("Yellow")              #tuple with only 1 element is not a tuple but a STRING
print(type(T))
T = ("Yellow",)              #tuple with 1 element with coma(,)
print(type(T))
```

```python
#accessing elements in tuples
T = ("Yellow","Green","Orange","Red","Black","Purple")
print(T[1])
print(T[-1])
print(T[2:4])
```

```python
#Remove
T = ("Yellow","Green","Orange")
print(T)
L = list(T)
L.remove("Yellow")
T = tuple(L)
print(T)
```

```python
#deleting tuple
T = ("Yellow","Green","Orange")
del(T)
#print(T) #printing after deletion is not possible#
```

```python
#appending element in tuple
T = ("Yellow","Green","Orange")
print(T)
L = list(T)
L.append("Grey")
T = tuple(L)
print(T)
```
