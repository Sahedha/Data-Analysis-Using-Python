```python
#There are 5 subjects psycology,nano technology,medicine,Anchology,anthrology 2 students have applied to do masters in ANY OF THESE 5 fields with alineplot show in which field they can excel
#the cuttoff for each subject marks is 70  student1 [67,45,73,71,57]   std2[43,67,68,53,85]
sub=["psy","nano","Med","Onc","Anthro"]
S1=[67,45,73,71,57]
S2=[43,67,68,53,85]
cut=[70,70,70,70,70]
#student1
#syntax=plt.plottype(x,y,label,colour,marker,linewidth,linestyle)
plt.scatter(sub,S1,label='Student1',color="k",marker="P")
#student2
plt.scatter(sub,S2,label="Student2",color="b",marker="h")
#cutoff
plt.plot(sub,cut,label="Cutoff",color="r",linestyle=":",linewidth=4)
plt.title("Student marks ")
plt.xlabel('Subject values')
plt.ylabel('Marks values')
plt.legend()
```

```python
#An India VS Australia match is online.India has given 298 as target for australia,The runs for every 5overs is given as follows [25,51,84,131,160,189,220,250,267,297]
#Australia has completed upto 25 0vers runs of aus[15,41,94,110,151]
#plot OVERS vs RUNS for both the countries with Ind has blue line and Aus has yellow with required labels and legends the marker for Ind is Flag">" and Aus is a "+"Take overs from "arange" function
#create arrays from every run and linebit=2
import numpy as np
Ind = np.array([25,51,84,131,160,189,220,250,267,297])
Aus = np.array([15,41,94,110,151])
IndOvers = np.arange(5,51,5)
AusOvers = np.arange(5,26,5)
#Indovers
plt.plot(IndOvers,Ind,label='India',color="b",marker=">",linewidth=2)
#AusOvers
plt.plot(AusOvers,Aus,label='Australia',color="y",marker="P",linewidth=2)
plt.title("India vs Australia")
plt.xlabel('Overs')
plt.ylabel('Runs')
plt.legend()
plt.grid()
```
