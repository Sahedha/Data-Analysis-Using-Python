```python
#Mean finding mean in a field
Mydf = pd.read_csv("/content/grades_withnulls.csv")
Mv = Mydf['SEM1'].mean()              #finding mean in SEM1
print(Mv)
Mydf.fillna(Mv,inplace = True)           #filling the coming mean with null values and saving the changes
Mydf
```

Accessing your data

```python
#iloc  = access integer location only -index
#loc  = access field names,index                record is a row && field is a column
#syntax: dfname.loc[index]----->rows
#syntax: dfname.loc[start:stop]--->range of rows
#dfname.loc[r_ind,col_name]---->rows &columns
#dfname.loc[]
Mydf.loc[0]      #first record
```

```python
Mydf.loc[0:5]     #first 5 records can give any range of numbers
```

```python
#mentioning column name      -----loc acepts both index and names
Mydf.loc[8:10,"SEM1"]
```

```python
Mydf.iloc[10:12,0:2]     #both rows and columns here mentioned in form of index    ----iloc only acepts index
```

```python
#accessing only specific data
Mydf[Mydf.SEM2==9.1]
```

```python
#print records who scored 9.3 in sem3
Mydf[Mydf.SEM3==9.3]
```

```python
#range of values
Mydf.loc[Mydf["SEM1"] > 9.8 , ["Names"]]
```

```python
#access and print the grades of  students who scored more than 9.7 in sem3
Mydf.loc[Mydf["SEM3"] > 9.7 ,"Grade"]
#accessing the records with sem39.7 but printing only Grade column
```

```python
#accessing using index function
print(Mydf[Mydf.index==7])
print(Mydf.loc[Mydf.index==7,"Grade"])
```

```python
#removing duplicates
Mydf.drop_duplicates()
```

```python
Mydf.drop_duplicates(inplace=True)
Mydf                      #to save the changes
```
