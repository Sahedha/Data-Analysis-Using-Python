Extracting HTML code of a Web page

-Program flow

*   Import libraries
*   Save the URL
*   Using requests.get(),access the web page
*   Using BeautifulSoup(),access the HTML code



```python
from bs4 import BeautifulSoup as bs
```

```python
import requests as rq
```

```python
url = "https://www.webtoons.com/en/"
```

```python
page = rq.get(url)
```

```python
code = bs(page.text,"html")
print(code)
```

```python
t = code.find('title')
print(t)
```

```python
print(t.text)
```

```python
url1 = "https://www.forbesindia.com/article/explainers/top-10-richest-people-india/85909/1"
```

```python
page1 = rq.get(url1)
```

```python
soup = bs(page1.text,"html")
print(soup)
```

```python
tab = soup.find("table")
print(tab)
```

```python
head = tab.find_all("th")
print(head)
```

```python
Heads = [i.text for i in head]
print(Heads)
```

```python
import pandas as pd
```

```python
#in columns only text
mydf = pd.DataFrame(columns=Heads)
print(mydf)
```

```python
#data
rows = tab.find_all("tr")
print(rows)
```

```python
#table data
td = tab.find_all("td")
print(td)
```

```python
row = [i.text for i in rows]
print(row)
```

```python
tabdata1 = [i.text for i in td]
print(tabdata1)
```

```python
tabdata1[0:4]
```

```python
mydf1 = pd.DataFrame(columns=["Name","Rank","NW","Company"])
print(mydf1)
```

```python
srt = 0
stp = 4
ind=0
for i in tabdata1:
  if stp <= 40:
    mydf1.loc[ind] = tabdata1[srt : stp]
    srt =srt+4
    stp =stp+4
  ind=ind+1
print(mydf1)
```

```python
mydf1
```

```python
mydf1.to_csv("mydfram1.csv")
```
