```python
#Take a number from the user which is a salary,If the salary is more then 50000 there is a 10% tax print his final salary.
a=int(input("Enter Salary"))
if(a>=50000):
  print("The salary after deducting tax: ",a-(a*(10/100)))
else:
  print("Your salary is less than 50000 So no tax deduction")
```

```python
#Get the salary from the user,if the salary is more than or equal to 50k apply 10% tax deduction ,if the salary is less than 50k apply 5% tax, if the salary is more than 70k apply 12% tax, if your salary exceeds 1L print Your promoted to next level
sal=int(input("Enter Salary: "))
if(sal<=50000):
  print("The salary after deducting tax 5%: ",sal-(sal*(5/100)))
elif(sal<=70000):
  print("The salary after deducting tax 10%: ",sal-(sal*(10/100)))
elif(sal<=100000):
  print("The salary after deducting tax 12%: ",sal-(sal*(12/100)))
else:
  print("You are Promoted")
```

```python
#create a 1d array
import numpy as np
a=np.array([1,2,3,4])        #1d array
print(a.ndim)
print(a.shape)
```

```python
#create a 2 dim array
import numpy as np
a=np.array([[1,2,3,4],[5,6,7,8],[15,66,77,88]])
#element in each row=coloumn                              #2d array
print(a.ndim)
print(a.shape)
print(a)
```

```python
#create a 3d array
import numpy as np
a=np.array([[[1,2,3,4],[5,6,7,8],[15,66,77,88]],[[1,2,9,8],[1,3,4,3],[5,6,5,9]]])
#element in each row=coloumn                              #3d array
print(a.ndim)
print(a.shape)
print(a)
```
