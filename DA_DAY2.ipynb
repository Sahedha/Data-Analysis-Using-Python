```python
A = np.array([35,56,78,90]).reshape(2,2)
B = np.array([7,9,3,12]) .reshape(2,2)
print(A)
print("____________________")
print(B)
print("____________________")
C = A*B                                #one on one mul in matrix
print(C)
```

```python
A = np.array([35,56,78,90]).reshape(2,2)
B = np.array([7,9,3,12]) .reshape(2,2)
print(A)
print("____________________")
print(B)
print("____________________")
C = A@B
D = A.dot(B)                      #Matrix multiplication
print(C)
print("____________________")
print(D)
```

```python
#min max of every rows and columns in array
A = np.array([35,56,78,90,632,869,12,8,221,45,27,123,75,987,34,20]).reshape(4,4)
print(A)
print("____________________")
print("max in col",A.max(axis=0))             #axis=0 i.e.,column
print("____________________")
print("max in row",A.max(axis=1))             #axis=1 i.e.,row
print("____________________")
print("min in col",A.min(axis=0))
print("____________________")
print("min in row",A.min(axis=1))
```

```python
#joining arrays  in vertically stacking                      vstac --- axis=0
A = np.array([35,56,78,90]).reshape(2,2)
B = np.array([7,9,3,12]) .reshape(2,2)
print(A)
print("__________________________")
print(B)
print("After vertically stacking two arrays")
print(np.vstack((A,B)))                                 #if (B,A) then B matrix will be on top
```

```python
#joining 2 matrix horizontally stacking                      hstack --- axis=1
A = np.array([35,56,78,90]).reshape(2,2)
B = np.array([7,9,3,12]) .reshape(2,2)
print(A)
print("__________________________")
print(B)
print("After horizontally stacking two arrays")
print(np.hstack((A,B)))                                 #if (B,A) then B matrix will be on front
```

```python
 #dstack                                                according to dimension
A=np.arange(30).reshape(2,3,5)
print(A)
print("After dstack")
print(np.dstack(A)) #no.of rows --->  no.of groups
#no.of columns in every group --->  no.of rows in every group
```

```python
#Splitten array
A=np.arange(18).reshape(6,3)
print(A)
print("After splitting vertically")
np.vsplit(A,3)                          #(A,num) num = no.of matrices we are going to split
```

```python
#Splitten array in our way
A=np.arange(18).reshape(6,3)
print(A)
print("After splitting vertically")
np.vsplit(A,(2,3,5)  )
```

```python
#splitting horizontally array
A=np.arange(18).reshape(3,6)
print(A)
print("After splitting horizontally")
np.hsplit(A,3)
```

```python
#splitting horizontally array in our choice
A=np.arange(18).reshape(3,6)
print(A)
print("After splitting horizontally")
np.hsplit(A,(2,5))
```

```python
#argmax in column ||  a.argmax()  if max in whole array             argmax is frequently used in pixels
A = np.array([17,4,18,56,12,33,89,21,34,55,11,13]).reshape(3,4)                #[0 0 0]
print(A)                                                                       #[1 1 1]
print(A.argmax(axis = 0))                                                      #[2 2 2]
```

```python
#argmax of rows in array
A = np.array([17,4,18,56,12,33,89,21,34,55,11,13]).reshape(3,4)             #[[0 1 2]
print(A)                                                                    # [0 1 2]
print(A.argmax(axis = 1))                                                   # [0 1 2]]
```

```python
#argmin of rows,columns in arrays
A = np.array([17,4,18,56,12,33,89,21,34,55,11,13]).reshape(3,4)
print(A)
print("columns",A.argmin(axis = 0))
print("_______________")
print("rows",A.argmin(axis = 1))

```
